```python
movie = df1['IMDB_Rating'].value_counts()

f=plt.figure(figsize = (10, 5))
plt.bar( movie.keys(),movie, color ='blue', width = 0.2) 
plt.xlabel("Rating") 
plt.ylabel("No Of Movies") 
plt.title("Rating VS No of Movies") 
plt.show()
```
